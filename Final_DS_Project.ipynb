```python
class FoodItem:
    def __init__(self, name, quantity, price, discount, stock):
        self.food_id = None
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

food_items = []

def generate_food_id():
    if len(food_items) == 0:
        return 1
    else:
        return food_items[-1].food_id + 1

def add_food_item(name, quantity, price, discount, stock):
    food_id = generate_food_id()
    food_item = FoodItem(name, quantity, price, discount, stock)
    food_item.food_id = food_id
    food_items.append(food_item)
    return food_item

def edit_food_item(food_id, name, quantity, price, discount, stock):
    for food_item in food_items:
        if food_item.food_id == food_id:
            food_item.name = name
            food_item.quantity = quantity
            food_item.price = price
            food_item.discount = discount
            food_item.stock = stock
            return food_item
    return None

def view_food_items():
    return food_items

def remove_food_item(food_id):
    for food_item in food_items:
        if food_item.food_id == food_id:
            food_items.remove(food_item)
            return True
    return False

class User:
    def __init__(self, data):
        self.data = data
        self.orders = []

    def place_new_order(self, menu):
        print("Available Food Menu:")
        for i, item in enumerate(menu, start=1):
            print(f"{i}. {item['name']} [{item['description']}] [INR {item['price']}]")

        selected_items = []
        while True:
            try:
                selections = input("Enter the item numbers you want to order (comma-separated): ").split(',')
                selections = [int(item.strip()) for item in selections]
                for sel in selections:
                    if 1 <= sel <= len(menu):
                        selected_items.append(menu[sel - 1])
                break
            except ValueError:
                print("Invalid input. Please enter valid item numbers.")

        self.orders.append(selected_items)
        print("Items selected for order:")
        for item in selected_items:
            print(f"{item['name']} [{item['description']}] [INR {item['price']}]")
        confirm_order = input("Do you want to confirm this order? (yes/no): ").strip().lower()
        if confirm_order == "yes":
            total_price = sum(item['price'] for item in selected_items)
            print(f"Order placed successfully. Total Price: INR {total_price}")
        else:
            print("Order canceled.")

    def view_order_history(self):
        if not self.orders:
            print("No order history available.")
        else:
            print("Order History:")
            for i, order in enumerate(self.orders, start=1):
                print(f"Order {i}:")
                for item in order:
                    print(f"{item['name']} [{item['description']}] [INR {item['price']}]")
                print()

    def update_profile(self):
        print("Update Profile:")
        self.data["full_name"] = input("Full Name: ")
        self.data["phone_number"] = input("Phone Number: ")
        self.data["email"] = input("Email: ")
        self.data["address"] = input("Address: ")
        self.data["password"] = input("Password: ")
        print("Profile updated successfully!")

# Function to register a new user
def register_user():
    print("User Registration:")
    full_name = input("Full Name: ")
    phone_number = input("Phone Number: ")
    email = input("Email: ")
    address = input("Address: ")
    password = input("Password: ")

    # Create a new user data entry
    user_data[email] = {
        "full_name": full_name,
        "phone_number": phone_number,
        "email": email,
        "address": address,
        "password": password,
        "role": "user"
    }
    print("Registration successful!")

# Define a dictionary to store user data
user_data = {
    "admin@example.com": {
        "full_name": "Admin",
        "phone_number": "1234567890",
        "email": "admin@example.com",
        "address": "Admin Address",
        "password": "admin123",
        "role": "admin"
    }
}

# Sample food menu
food_menu = [
    {"name": "Tandoori Chicken", "description": "(4 pieces)", "price": 240},
    {"name": "Vegan Burger", "description": "(1 piece)", "price": 320},
    {"name": "Truffle Cake", "description": "(500gm)", "price": 900}
]

def main():
    print("Welcome to the Food Ordering App!")

    while True:
        print("\nMain Menu:")
        print("1. Log in")
        print("2. Register")
        print("3. Exit")
        choice = input("Enter your choice: ")
        if choice == '1':
            email = input("Email: ")
            password = input("Password: ")
            if email in user_data and user_data[email]["password"] == password:
                user = User(user_data[email])
                if user.data["role"] == "admin":
                    print("Admin login successful.")
                    while True:
                        print("\nAdmin Menu:")
                        print("1. Add new food item")
                        print("2. Edit food item")
                        print("3. View all food items")
                        print("4. Remove food item")
                        print("5. Log out")
                        admin_choice = input("Enter your choice: ")
                        if admin_choice == '1':
                            name = input("Enter food item name: ")
                            quantity = input("Enter quantity (e.g., 100ml, 250gm, 4 pieces): ")
                            price = float(input("Enter price: "))
                            discount = float(input("Enter discount: "))
                            stock = int(input("Enter stock amount: "))
                            add_food_item(name, quantity, price, discount, stock)
                            print("Food item added successfully!")
                        elif admin_choice == '2':
                            food_id = int(input("Enter FoodID to edit: "))
                            food_item = edit_food_item(food_id, input("Enter new name: "), input("Enter new quantity: "),
                                                       float(input("Enter new price: ")), float(input("Enter new discount: ")),
                                                       int(input("Enter new stock amount: ")))
                            if food_item:
                                print("Food item updated successfully!")
                            else:
                                print("Food item not found.")
                        elif admin_choice == '3':
                            print("\nList of all food items:")
                            for item in view_food_items():
                                print(f"FoodID: {item.food_id}, Name: {item.name}, Quantity: {item.quantity}, Price: {item.price}, Discount: {item.discount}, Stock: {item.stock}")
                        elif admin_choice == '4':
                            food_id = int(input("Enter FoodID to remove: "))
                            if remove_food_item(food_id):
                                print("Food item removed successfully!")
                            else:
                                print("Food item not found.")
                        elif admin_choice == '5':
                            print("Logging out.")
                            break
                        else:
                            print("Invalid choice. Please try again.")
                else:
                    print("User login successful.")
                    while True:
                        print("\nUser Menu:")
                        print("1. Place New Order")
                        print("2. Order History")
                        print("3. Update Profile")
                        print("4. Log Out")
                        user_choice = input("Enter your choice: ")
                        if user_choice == '1':
                            user.place_new_order(food_menu)
                        elif user_choice == '2':
                            user.view_order_history()
                        elif user_choice == '3':
                            user.update_profile()
                        elif user_choice == '4':
                            print("Logging out.")
                            break
                        else:
                            print("Invalid choice. Please try again.")
            else:
                print("Invalid email or password. Please try again.")
        elif choice == '2':
            register_user()
        elif choice == '3':
            print("Exiting the application.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
